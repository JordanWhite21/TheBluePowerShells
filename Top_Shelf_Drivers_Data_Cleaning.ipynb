```python
import requests
import pandas as pd
```

```python
AllCourses = "https://mario-kart-tour-api.herokuapp.com/api/v1/courses/"
```

```python
# Gather the course data
CourseData = requests.get(AllCourses).json()
```

```python
# LEVEL ONE TOP SHELF DRIVER CLEANING 
```

```python
# For each course, collect the level one top sheld driver IDs
level_1_top_shelf_driver = {}
for course in CourseData:
    value = []
    for driver in course["level_one_top_shelf_drivers"]:
        value.append(driver["id"])
    level_1_top_shelf_driver[course["name"]]=value
level_1_top_shelf_driver
```

```python
# Create a dataframe
level_one_df = pd.DataFrame.from_dict(level_1_top_shelf_driver, orient='index')
level_one_df
```

```python
# Create a dataframe
level_one_df = pd.DataFrame.from_dict(level_1_top_shelf_driver, orient='index')
level_one_df
```

```python
# Make dataframe prettier
level_one_df = level_one_df.rename_axis("course_name")
level_one_df = level_one_df.rename(columns={0:"driver_id_1",1:"driver_id_2",2:"driver_id_3",3:"driver_id_4",4:"driver_id_5",5:"driver_id_6",6:"driver_id_7",7:"driver_id_8",8:"driver_id_9",9:"driver_id_10"})
level_one_df
```

```python
# Count NaN values per column
level_one_df.isna().sum()
```

```python
# Drop rows with NaN in driver_id_1,2,3,4,5 and check that they were dropped
level_one_df = level_one_df.dropna(subset=["driver_id_1","driver_id_2","driver_id_3","driver_id_4","driver_id_5"])
level_one_df.isna().sum()
```

```python
# Drop remaining columnds with NaN values
level_one_df = level_one_df.drop(columns=["driver_id_6","driver_id_7","driver_id_8","driver_id_9","driver_id_10"])
level_one_df
```
