```python
import requests
import pandas as pd
import os
```

```python
AllCourses = requests.get("https://mario-kart-tour-api.herokuapp.com/api/v1/courses/")
AllDrivers = requests.get("https://mario-kart-tour-api.herokuapp.com/api/v1/drivers/")
```

```python
#Gather the course data
#Request into JSON
CourseData = AllCourses.json()
# JSON Data to Dataframe
courses_df = pd.DataFrame(CourseData)
#Show the DF
courses_df
```

```python
#Create our DriverDF
#Gather the Driver data
DriverData = AllDrivers.json()
drivers_df = pd.DataFrame(DriverData)
drivers_df
```

```python
#Check the datatypes
drivers_df.dtypes
```

```python
#Cleaned Driver DF
cleaned_drivers_df = drivers_df.drop(['debut_tour','date_added','level_one_favorite_courses','level_three_favorite_courses','level_six_favorite_courses','is_daily_select',
                                      'is_in_pipes','baby','crown','dress','earrings','extended_tongue','gloves','helmet','horns','koopaling','mustache',
                                     'ribbon','shell','short_sleeves','three_hairs','tie','hat','kong'],axis=1)
cleaned_drivers_df
```

```python
#export the cleaned_df to CSV
cleaned_drivers_df.to_csv("/Resources/drivers")
```
