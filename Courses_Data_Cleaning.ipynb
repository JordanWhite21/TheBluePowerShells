```python
import requests
import pandas as pd
```

```python
AllCourses = "https://mario-kart-tour-api.herokuapp.com/api/v1/courses/"
```

```python
# Gather the course data
CourseData = requests.get(AllCourses).json()

# Create our DataFrame
courses_df = pd.DataFrame(CourseData)
courses_df
```

```python
# Drop unwanted columns: debut_tour, debut_game, debut_system, date_added, is_ranked_week_1, is_ranked_week_2
drop_columns = ["debut_tour","debut_game","debut_system","date_added","is_ranked_week_1","is_ranked_week_2"]
courses_df = courses_df.drop(columns=(drop_columns), axis=1)
courses_df
```

```python
# Create a dataframe for level_one_top_shelf_drivers
level_1_drivers_df = pd.DataFrame().assign(level_1=courses_df["level_one_top_shelf_drivers"][0])
level_1_drivers_df
```

```python
# See what's available in the list of dictionaries for level_one_drivers_df
level_1_drivers_df["level_1"][0]
```

```python
# Call each ID 
level_1_drivers_df["level_1"][0]["id"]
level_1_drivers_df["level_1"][1]["id"]
level_1_drivers_df["level_1"][2]["id"]
```

```python
# Attempt to pluck out ID
level_1_drivers_df = pd.DataFrame().assign(driver_1=level_1_drivers_df["level_1"][0]["id"])
level_1_drivers_df
```

```python
# Create a dataframe for level_one_top_shelf_drivers
level_3_drivers_df = pd.DataFrame().assign(Level3=courses_df["level_three_top_shelf_drivers"])
level_3_drivers_df
```

```python
# Create a dataframe for level_one_top_shelf_drivers
level_6_drivers_df = pd.DataFrame().assign(Level6=courses_df["level_six_top_shelf_drivers"])
level_6_drivers_df
```

```python
# Drop level_one_top_shelf_drivers, level_three_top_shelf_drivers, level_six_top_shelf_drivers from courses_df
drop_levels = ["level_one_top_shelf_drivers", "level_three_top_shelf_drivers", "level_six_top_shelf_drivers"]
courses_df = courses_df.drop(columns=(drop_levels))
courses_df
```

```python
# Export to courses_df to CSV
courses_df.to_csv("courses.csv", index=False)
```
